```python
#import dependencies
import pandas as pd
import numpy as np
```

```python
#import csv
"""df = pd.read_csv("Resources/train.csv")
 got this warning:
 DtypeWarning: Columns (26) have mixed types. Specify dtype option on import or set low_memory=False."""

df = pd.read_csv("Resources/train.csv", dtype={26: str})
```

```python
#check for inital duplicates
duplicates = df.duplicated().sum()
print(f"The number of duplicates is {duplicates}")
```

```python
#check info types for correct format
df.info()
```

```python
#check shape
df.shape
```

```python
""" dont want: id, cust id, month, name, age, social occupation, no. bank accounts, interst rate, invest monthly  """
#dropping unwanted columns
# dropped_df = df.drop(['ID','Customer_ID','Month','Name', 'Age', 'Occupation', 'Num_Bank_Accounts', 'Interest_Rate', 'Amount_invested_monthly'], axis = 1)
dropped_df = df.drop(['ID','Customer_ID','Month','Name'], axis = 1)
#formatting spaces
df = dropped_df.replace("_", " ", regex=True)
#fill in blank cells
#refrenced from stack overflow
df.replace(r'^\s+$', np.nan, regex=True)
df
```

```python
#dropping negative values
columns_to_check = [
    'Annual_Income', 'SSN', 'Monthly_Inhand_Salary', 'Num_Credit_Card',
    'Num_of_Loan', 'Delay_from_due_date', 'Num_of_Delayed_Payment',
    'Changed_Credit_Limit', 'Num_Credit_Inquiries', 'Outstanding_Debt',
    'Credit_Utilization_Ratio', 'Total_EMI_per_month', 'Monthly_Balance'
]

# Convert specified columns to numeric, setting errors='coerce' to turn non-numeric values into NaN
df[columns_to_check] = df[columns_to_check].apply(pd.to_numeric, errors='coerce')

# Drop rows where any of the specified columns have negative values
df.drop(df[df[columns_to_check].lt(0).any(axis=1)].index, inplace=True)

#dropping rows which have the social security number error "#F%$D@*&8"
#df = df.drop(df[df['SSN'] == '#F%$D@*&8'].index)

df.shape
```
